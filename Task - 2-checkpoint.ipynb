```python
# Importing all required library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load dataset
df = pd.read_csv("train.csv")  # or whatever name you used
df.head()
```

```python
# Univariate Analysis
# Visualizing distributions with histograms

# Histograms are great for spotting skewness and distribution shape
df.hist(figsize=(15, 10), bins=15, color='purple')
plt.suptitle("Histogram of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Univariate Analysis
# Boxplots for Outlier Detection

# Plot boxplots for each numeric column
numerical_cols = df.select_dtypes(include=np.number).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis
# Correlation Heatmap

# Select only numerical columns for correlation
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap (Numerical Features Only)", fontsize=16)
plt.show()
```

```python
# Bivariate Analysis
# Pairplot of Selected Features with Target

# Only using few columns to avoid clutter
selected = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Loan_Status']
sns.pairplot(df[selected], hue="Loan_Status")
plt.show()
```

```python
# Skewness of Numerical Features

# Detect skewed features that may need transformation
print("Skewness of Numerical Features:")
print(df[numerical_cols].skew())
```

```python
# Multicollinearity - Highly Correlated Features

# Get correlation matrix
corr_matrix = df[numerical_cols].corr().abs()

# Take upper triangle of matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Identify features with correlation greater than 0.75
high_corr = [col for col in upper.columns if any(upper[col] > 0.75)]
print("Highly correlated features (possible multicollinearity):", high_corr)
```

```python
# Bivariate Analysis - Categorical vs Target (Loan_Status)

# Education vs Loan_Status
sns.countplot(x='Education', hue='Loan_Status', data=df, palette='Set2')
plt.title("Loan Approval by Education")
plt.show()

# Property_Area vs Loan_Status
sns.countplot(x='Property_Area', hue='Loan_Status', data=df, palette='Set1')
plt.title("Loan Approval by Property Area")
plt.show()

# Credit_History vs Loan_Status
sns.countplot(x='Credit_History', hue='Loan_Status', data=df, palette='Set3')
plt.title("Loan Approval by Credit History")
plt.show()
```

```python
# Bivariate Analysis - Numerical vs Target (Loan_Status)

# ApplicantIncome vs Loan_Status
sns.boxplot(x='Loan_Status', y='ApplicantIncome', data=df, palette='pastel')
plt.title("Applicant Income by Loan Status")
plt.show()

# LoanAmount vs Loan_Status
sns.boxplot(x='Loan_Status', y='LoanAmount', data=df, palette='coolwarm')
plt.title("Loan Amount by Loan Status")
plt.show()
```
